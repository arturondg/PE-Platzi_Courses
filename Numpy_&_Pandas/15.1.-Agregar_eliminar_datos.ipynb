# Agregar o eliminar datos con Pandas

```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('/home/arturo/Practice_Projects/Curso_Numpy_Pandas/13.2.-bestsellers-with-categories.csv',sep=',', header=0,names=['Nombres', 'Autor', 'Calificación', 'Reseña', 'Precio', 'Año', 'Género'])
```

```python
#Head muesta el encabezado con las filas especificadas
df_books.head(2)
```

### •Drop columns

```python
#Drop elimina el dato de salida
df_books.drop('Género',axis=1).head(2)
```

```python
#Pero no se elimina permanentemente al traerlos de nuevo
df_books.head(2)
```

#### +Borrado permanente

##### -Con inplace

```python
#Se hace con inplace
df_books.drop('Género',axis=1,inplace=True)
```

```python
#El borrado será permanente en salida
df_books.head(2)
```

##### -Con variable

```python
#Si se hace una nueva variable también hay borrado
df_books = df_books.drop('Año',axis=1)
```

```python
df_books.head(2)
```

##### -Con del

```python
#del es una función de Python, y hace casi lo mismo(No recomendada)
del df_books['Precio']
```

```python
df_books.head(2)
```
